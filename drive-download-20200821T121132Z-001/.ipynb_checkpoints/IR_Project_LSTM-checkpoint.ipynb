```
newArticle=[]
for x in articles.tolist():
  x=x.lower()
  #print("before:",x)
  for word in STOPWORDS:
    token = ' ' + word + ' '
    x = x.replace(token, ' ')
    #x = x.replace(' ', ' ')
  newArticle.append(x)
  #print("after:",x)
print(len(newArticle))
```

```
R_newArticle=[]
for x in newArticle:
  y=x.split(' ', 1)
  if y[0] in STOPWORDS:
    R_newArticle.append(y[1])
  else:
    R_newArticle.append(x)
#print(R_newArticle[706])
newArticle=R_newArticle
```

```
def BIlstm():
 model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(embedding_dim,dropout=0.2)),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
 ])
  #model.summary()
 return model

```

```
def BIlstm2():
 embedding_dim=100
 model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(embedding_dim,dropout=0.4)),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
 ])
  #model.summary()
 return model
```

```
def BIlstm3():
 embedding_dim=100
 model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(embedding_dim)),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(50, activation='relu'),
    tf.keras.layers.Dropout(0.4),
    tf.keras.layers.Dense(4, activation='softmax')
 ])
  
 return model
```

```
def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.plot(history.history['val_'+string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, 'val_'+string])
  plt.show()
```

```
def LSTM():
 model2 = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim),
    tf.keras.layers.LSTM(embedding_dim,dropout=0.2),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
 ])
 #model2.summary()
 return model2
```

```
def GRU(): 
  model3 = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim),
    tf.keras.layers.GRU(embedding_dim,dropout=0.2),
    tf.keras.layers.Dense(30, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
  ])
  #model2.summary()
  return model3
```

```
def runMyModel(model,epochs,batchSize,pat):
  callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=pat,verbose=0)
  model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
  num_epochs = epochs
  batch=batchSize
  history = model.fit(train_padded, training_label_seq, epochs=num_epochs, batch_size=batch,
                      callbacks=[callback],validation_data=(validation_padded, validation_label_seq))
  ypredict = np.argmax(model.predict(test_padded), axis=-1)
  ytest=[]
  for x in test_label_seq:
    ytest.append(int(x))
  target_names=['none','bjp', 'congress']
  print(classification_report(ytest, ypredict, target_names=target_names))
  print("F1-score:",f1_score(ytest, ypredict,average='weighted'))
  print("Accuracy:",accuracy_score(ytest, ypredict))
```

```
batchSize=[5,10,15,20,30,40]
epochs=[5,10,15,20,25]
for x in batchSize:
  for y in epochs:
    model=LSTM()
    runMyModel(model,y,x)
```

```
batchSize=[10,15,20,30,40]
epochs=[5,10,15,20,25]
for x in batchSize:
  for y in epochs:
    model=GRU()
    runMyModel(model,y,x)
```
