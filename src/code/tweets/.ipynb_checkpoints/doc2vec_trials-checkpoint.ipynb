```python
k = ["fauci"]
import GetOldTweets3 as got
import csv
from datetime import datetime
def query(target_csv, keywords, accounts_file, date_since):
    pd.set_option('mode.chained_assignment', None)
    df = pd.read_csv(accounts_file)
    cw = csv.writer(open(target_csv, "w"))
    new_df = pd.DataFrame({
        'name':[], 'twitter':[], 'party':[], 'tweet':[], 'tweet_published':[]
    })

    for index, row in df.iterrows():
        try:
            # print(index, row['name'], row['twitter'], row['party'])
            tweets_set = set()
            for word in keywords:
                tweet =  got.manager.TweetCriteria().setUsername(row['twitter'])\
                                                        .setSince(date_since)\
                                                        .setUntil(datetime.now().strftime("%Y-%m-%d"))\
                                                        .setQuerySearch(word)
                tweetObj = got.manager.TweetManager.getTweets(tweet)
                for tw in tweetObj:
                    tweets_set.add((tw.text, tw.date.strftime("%Y-%m-%d")))
            # print(tweets_hashmap)
            print(row['twitter'])

            for tweet in tweets_set:
                new_row  = {'name': row['name'], 'twitter': row['twitter'], 'party': row['party'], 'tweet': tweet[0],'tweet_published': tweet[1]}
                new_df = new_df.append(new_row, ignore_index=True)
        except:
            continue
    return new_df
d = query("tweets.csv", k, twitter_url_reps, "2019-12-25")
d
```

```python
corpus = [x["tweet"] for i,x in d.iterrows()]
```

```python
import gensim
def read_corpus(corpus, tokens_only=False):
    for i, line in enumerate(corpus):
        tokens = gensim.utils.simple_preprocess(line)
        if tokens_only:
            yield tokens
        else:
            # For training data, add tags
            yield gensim.models.doc2vec.TaggedDocument(tokens, [i])
            
train_corpus = list(read_corpus(corpus))
test_corpus = list(read_corpus(corpus, tokens_only=True))
train_corpus[:2]
```

```python
len(train_corpus)
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=200, min_count=2, epochs=1000)
model.build_vocab(train_corpus)
```

```python
model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
v1 = model.infer_vector(train_corpus[0].words)
print("Vector of tweet \n {} \n is {}".format(corpus[0], v1))
```

```python
import torch
import torch.nn as nn
```

```python
device = torch.device("cuda")
```

```python
classifier = nn.Sequential(
    nn.Linear(200, 400),
    nn.Sigmoid(),
    nn.Linear(400, 2),
    nn.Sigmoid(),
    nn.Softmax(dim = 0)
)
```

```python
import numpy as np
lr = 0.1
loss_list = list([])
graphable_loss = []
for i in range(0,100000):
    tweetidx = np.random.randint(low = 0, high = 430)
    v = torch.tensor(model.infer_vector(train_corpus[tweetidx].words), dtype = torch.float)
    real = torch.tensor([0.0,1.0] if d.at[tweetidx, "party"] == "Republican" else [1.0, 0.0], dtype = torch.float) 
    pred = classifier(v)
    loss = (pred-real).pow(2).sum()
    loss_list.append(loss)
    if len(loss_list) > 10:
        loss_list.pop(0)
        if i % 100 == 0:
            acc = 0.0
            for l in loss_list:
                acc+= l
            graphable_loss.append(acc/10.0)
            print("Loss at trial {} is {}".format(i, acc/10.0))
    classifier.zero_grad()
    loss.backward()
    with torch.no_grad():
        for param in classifier.parameters():
            param -= lr * param.grad
```

```python
for i in range(0, len(graphable_loss)):
    graphable_loss[i] = graphable_loss[i].item()
```

```python
c_loss_data = {}
c_loss_data["Loss"] = graphable_loss
```
