```python
dem_tweets = [x for i,x in d.iterrows() if x["party"] == "Democrat"]
rep_tweets = [x for i,x in d.iterrows() if x["party"] == "Republican"]
print(len(rep_tweets))
print(len(dem_tweets))
```

```python
corpus = [x["tweet"] for i,x in d.iterrows()]
```

```python
import gensim
def read_corpus(corpus, tokens_only=False):
    for i, line in enumerate(corpus):
        tokens = gensim.utils.simple_preprocess(line)
        if tokens_only:
            yield tokens
        else:
            # For training data, add tags
            yield gensim.models.doc2vec.TaggedDocument(tokens, [i])
            
train_corpus = list(read_corpus(corpus))
test_corpus = list(read_corpus(corpus, tokens_only=True))
train_corpus[:2]
```

```python
len(train_corpus)
```

```python
model = gensim.models.doc2vec.Doc2Vec(vector_size=400, min_count=2, epochs=20)
model.build_vocab(train_corpus)
```

```python
model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
v1 = model.infer_vector(train_corpus[0].words)
print("Vector of tweet \n {} \n is {}".format(corpus[0], v1))
```

```python
import torch
import torch.nn as nn
```

```python
device = torch.device("cuda")
```

```python
classifier = nn.Sequential(
    nn.Linear(len(v1), 400),
    nn.Sigmoid(),
    nn.Linear(400, 2),
    nn.Softmax(dim = 0)
)
```

```python
import numpy as np
lr = 0.02
loss_list = list([])
graphable_loss = []
party_domain = ["Democrat", "Republican"]
tweetidx = 0
for i in range(0,200000):
    while d.at[tweetidx, "party"] != party_domain[i%2]:
        tweetidx = np.random.randint(low = 0, high = len(train_corpus))
    v = torch.tensor(model.infer_vector(train_corpus[tweetidx].words), dtype = torch.float)
    real = torch.tensor([0.0,1.0] if tweetidx%2 == 1 else [1.0, 0.0], dtype = torch.float) 
    pred = classifier(v)
    loss = (pred-real).pow(2).sum()
    loss_list.append(loss)
    if len(loss_list) > 10:
        loss_list.pop(0)
        if i % 100 == 0:
            acc = 0.0
            for l in loss_list:
                acc+= l
            graphable_loss.append(acc/10.0)
            print("Loss at trial {} is {}".format(i, acc/10.0))
    classifier.zero_grad()
    loss.backward()
    with torch.no_grad():
        for param in classifier.parameters():
            param -= lr * param.grad
```

```python
for i in range(0, len(graphable_loss)):
    try:
        graphable_loss[i] = graphable_loss[i].item()
    except:
        continue
```

```python
c_loss_data = {}
c_loss_data["Loss"] = graphable_loss
```

```python
loss_df = pd.DataFrame(c_loss_data)
```
