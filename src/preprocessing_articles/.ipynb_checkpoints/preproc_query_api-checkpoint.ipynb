```python
equals = lambda b: (lambda a: a == b)
test_query = {}
cond1 = addCondition(test_query, "name", 10, equals("Hindustan Times"))
result = queryDB(test_query, 10000)
df = getParsedDF(result)
df
```

```python
greater = lambda b: (lambda a: a > b)
addSubCondition(cond1, "words_count", 5, greater(500))
result = queryDB(test_query, 10000)
df = getParsedDF(result)
df
```
