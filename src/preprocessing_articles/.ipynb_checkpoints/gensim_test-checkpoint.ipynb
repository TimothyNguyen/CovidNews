```python
import csv
```

```python
data_off_repo = "/home/data/covid/dataset/tweets1.csv"
# change this to some other static path pointing to off-repo path for dataset
```

```python
tweets = []
with open(data_off_repo, 'w', encoding='utf-8') as f:
        dict = csv.DictReader(f)
        for i,line in enumerate(dict):
            tweets.append({"party": line["party"], "tweet": line["tweet"]})
```

```python
for i in tweets:
    print(i["party"])
```
